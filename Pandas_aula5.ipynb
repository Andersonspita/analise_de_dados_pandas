```python
import pandas as pd
```

```python
#leitura de arquivos
df1 = pd.read_excel("Aracaju.xlsx")
df2 = pd.read_excel("Fortaleza.xlsx")
df3 = pd.read_excel("Natal.xlsx")
df4 = pd.read_excel("Recife.xlsx")
df5 = pd.read_excel("Salvador.xlsx")
```

```python
#juntando todos os arquivos
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
df.head(100)
```

```python
df.dtypes
```

```python
df["LojaID"] = df["LojaID"].astype("object")
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Consultando linhas com valores nulos
df.isnull().sum()
```

```python
#Substituindo valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
```

```python
#Apagando linhas com valores nulos
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#Apagando linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
df.isnull().sum()
```

```python
#removendo linhas com valores falyantes em todas as colunas
df.dropna(how="all", inplace=True)
```

```python
#Criando novas colunas
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#retornando maior receita
df["Receita"].max()
```

```python
#retornando menor receita
df["Receita"].min()
```

```python
#Top n maiores receitas
df.nlargest(3, "Receita")
```

```python
#top n menores receitas
df.nsmallest(3, "Receita")
```

```python
#Agrupando por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
df.sort_values("Receita", ascending=True).head(10)
```

```python
df.head()
```

```python
df["Data"] = df["Data"].astype("int64")
```

```python
df.dtypes
```

```python
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando nova coluna com ano
df["Ano_venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```
